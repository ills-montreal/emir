```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('modes_stats.csv')
```

```python


df['bic normalized'] = df.groupby(['model'])['bic'].transform(lambda x: (x - x.mean()) / x.std())
df['aic normalized'] = df.groupby(['model'])['aic'].transform(lambda x: (x - x.mean()) / x.std())

```

```python

fig, ax = plt.subplots()


sns.lineplot(data=df, x='modes', y='aic',hue='model',  legend=False, ax = ax, alpha=0.9)
# draw a vertical line at x=8
ax.axvline(x=8, color='r', linestyle='--')
```
